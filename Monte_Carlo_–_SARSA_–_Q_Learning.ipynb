```python
import numpy as np
import random
```

```python
grid_size = 4
start_state = (0, 0)
goal_state = (3, 3)

actions = [0, 1, 2, 3]
# 0: up, 1: down, 2: left, 3: right
```

```python
def step(state, action):
    x, y = state

    if action == 0:   # up
        x = max(x - 1, 0)
    elif action == 1: # down
        x = min(x + 1, grid_size - 1)
    elif action == 2: # left
        y = max(y - 1, 0)
    elif action == 3: # right
        y = min(y + 1, grid_size - 1)

    next_state = (x, y)

    if next_state == goal_state:
        reward = 1
    else:
        reward = -0.01

    return next_state, reward
```

```python
def epsilon_greedy(Q, state, epsilon):
    if random.random() < epsilon:
        return random.choice(actions)
    else:
        return np.argmax(Q[state])
```

```python
Q = {}
for i in range(grid_size):
    for j in range(grid_size):
        Q[(i, j)] = np.zeros(len(actions))
```

Monte Carlo Algorithm

```python
def monte_carlo(episodes, alpha=0.1, gamma=0.9, epsilon=0.2):

    for ep in range(episodes):
        state = start_state
        episode = []

        # Generate Episode
        while state != goal_state:
            action = epsilon_greedy(Q, state, epsilon)
            next_state, reward = step(state, action)
            episode.append((state, action, reward))
            state = next_state

        # Calculate Returns
        G = 0
        for state, action, reward in reversed(episode):
            G = gamma * G + reward
            Q[state][action] += alpha * (G - Q[state][action])
```

SARSA Algorithm (On-Policy)

```python
def sarsa(episodes, alpha=0.1, gamma=0.9, epsilon=0.2):

    for ep in range(episodes):
        state = start_state
        action = epsilon_greedy(Q, state, epsilon)

        while state != goal_state:
            next_state, reward = step(state, action)
            next_action = epsilon_greedy(Q, next_state, epsilon)

            Q[state][action] += alpha * (
                reward + gamma * Q[next_state][next_action] - Q[state][action]
            )

            state = next_state
            action = next_action
```

Q-Learning Algorithm (Off-Policy)

```python
def q_learning(episodes, alpha=0.1, gamma=0.9, epsilon=0.2):

    for ep in range(episodes):
        state = start_state

        while state != goal_state:
            action = epsilon_greedy(Q, state, epsilon)
            next_state, reward = step(state, action)

            best_next_action = np.max(Q[next_state])

            Q[state][action] += alpha * (
                reward + gamma * best_next_action - Q[state][action]
            )

            state = next_state
```

```python
monte_carlo(500)
sarsa(500)
q_learning(500)
```
